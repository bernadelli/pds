$$f(x) = cos(x)$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
pi = np.pi
x = np.linspace(-4*pi, 4*pi, 1000)

plt.plot(x,np.cos(x))
plt.title("f(x) = cos(x)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.show()
```

$$f(x) = sen(x)$$

```python
pi = np.pi
x = np.linspace(-4*pi, 4*pi, 1000)

plt.plot(x,np.sin(x))
plt.title("f(x) = sen(x)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.show()
```

$$f(x) = \Pi(x)$$

```python
def rect(x):
    tal = 1
    return np.where((abs(x)<(tal/2)),1.0,0.0)
```

```python
x = np.linspace(-2,2,500)

plt.plot(x,rect(x))
plt.title("f(x) = rect(x)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.show()
```

$$f(x)=u(x)$$

```python
def u(x):
    return np.where(x>0,1.0,0.0)
```

```python
x = np.linspace(-2,2,500)

plt.plot(x,u(x))
plt.title("f(x) = u(x)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.show()
```

$$f(x)=\Delta(x)$$

```python
def triang(x):
    tal = 1
    return np.where((abs(x)<(tal/2)),(1-(2*abs(x/tal))),0)
```

```python
x = np.linspace(-2,2,100)

plt.stem(x,triang(x))
plt.title("f(x) = triang(x)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.show()
```
