## Sinais cont√≠nuos

$$f(x) = cos(x)$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
pi = np.pi
x = np.linspace(-4*pi, 4*pi, 1000)

plt.plot(x,np.cos(x))
plt.title("f(x) = cos(x)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.show()
```

$$f(x) = sen(x)$$

```python
pi = np.pi
x = np.linspace(-4*pi, 4*pi, 1000)

plt.plot(x,np.sin(x))
plt.title("f(x) = sen(x)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.show()
```

$$f(x) = \Pi(x)$$

```python
def rect(x):
    tal = 1
    return np.where((abs(x)<(tal/2)),1.0,0.0)
```

```python
x = np.linspace(-2,2,500)

plt.plot(x,rect(x))
plt.title("f(x) = rect(x)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.show()
```

$$f(x)=u(x)$$

```python
def u(x):
    return np.where(x>0,1.0,0.0)
```

```python
x = np.linspace(-2,2,500)

plt.plot(x,u(x))
plt.title("f(x) = u(x)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.show()
```

$$f(x)=\Delta(x)$$

```python
def triang(x):
    tal = 1
    return np.where((abs(x)<(tal/2)),(1-(2*abs(x/tal))),0)
```

```python
x = np.linspace(-2,2,100)

plt.plot(x,triang(x))
plt.title("f(x) = triang(x)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.show()
```

$$f(x)=sinc(x)$$

```python
x = np.linspace(-4, 4, 500)

plt.plot(x,np.sinc(x))
plt.title("f(x) = sinc(x)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.show()
```

## Sinais discretos

```python
n = np.arange(-10., 11., 1.) #Defini o vetor n 
x = np.cos(pi * n / 4 ) #fun√ß√£o cosseno discreta

plt.stem(n,x)
plt.title("f[n] = cos($\Omega$ n)")
plt.xlabel("n")
plt.ylabel("f[n]")
plt.grid(True)
plt.xticks(n)
plt.xlim([ -10, 11 ])
plt.ylim([ -1.1, 1.1 ])
plt.show()
```

```python
n = np.arange(-10., 11., 1.) #Defini o vetor n 
x1 = np.cos(pi * n / 4 ) #fun√ß√£o cosseno discreta
x2 = np.sin(pi * n / 4 ) #fun√ß√£o cosseno discreta

fig, axs = plt.subplots(2, 1, constrained_layout=True)
fig.suptitle('f[n] = cos($\Omega$n) e f[n] = sen($\Omega$n)', color='blue',fontsize='large') 

axs[0].stem(n, x1)
axs[0].set_xticks(n)
#axs[0].set_xlim([ -10, 11 ])
#axs[0].set_ylim([ -1.1, 1.1 ])
axs[0].set_title('cos($\Omega$n)')
axs[0].set_xlabel('n')
axs[0].set_ylabel('f[n]')

axs[1].stem(n, x2)
axs[1].set_xticks(n)
axs[1].set_title('sen($\Omega$n)')
axs[1].set_xlabel('n')
axs[1].set_ylabel('f[n]')


plt.show()
```

$$f[n]=\delta[n]$$

```python
import scipy.signal as signal
imp = signal.unit_impulse(20, 'mid')

plt.stem(np.arange(-10, 10), imp)
plt.margins(0.1, 0.1)
plt.xlabel('Time [samples]')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

## Fun√ß√£o plotar

```python
tempo = np.linspace(-1, 1, 1000)
freq = 1 #Hz

def cos(tempo, freq):
    return signal.chirp(tempo, freq, 1, freq)
        
sinal = cos(tempo, freq)

def plotar(tempo, sinal, titulo = "", legendaX = "", legendaY = "", grade = True):
    plt.figure()
    plt.plot(tempo, sinal)
    plt.title(titulo)
    plt.xlabel(legendaX)
    plt.ylabel(legendaY)
    plt.grid(grade)
    
plotar(tempo, sinal, "Cossenoide, frequ√™ncia = 1Hz", "Tempo (segundos)")
```
